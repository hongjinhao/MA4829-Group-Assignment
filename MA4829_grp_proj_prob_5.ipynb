Onehotencode the data for apriori

```python
agedata = pd.DataFrame(personaliseddata)
#Append 4 columns for each age group
agedata["20-30"] = 0
agedata["31-40"] = 0
agedata["41-50"] = 0
agedata["51-60"] = 0

#Add 1 if age is 20-30
cond1 =agedata['age'] == '20-30'
agedata.loc[cond1,'20-30'] = 1
#Add 1 if age is 31-40
cond2 =agedata['age'] == '31-40'
agedata.loc[cond2,'31-40'] = 1
#Add 1 if age is 41-50
cond3 =agedata['age'] == '41-50'
agedata.loc[cond3,'41-50'] = 1
#Add 1 if age is 51-60
cond4 =agedata['age'] == '51-60'
agedata.loc[cond4,'51-60'] = 1
agedata
```

```python
#shorten design_comp answers
conddesign =agedata['design_comp'] == 'Only with professional help'
agedata.loc[conddesign,'design_comp'] = 'only pro'

#append design_comp categories
agedata["only pro"] = 0
agedata["Yes"] = 0
agedata["No"] = 0

condpro =agedata['design_comp'] == 'only pro'
agedata.loc[condpro,'only pro'] = 1

condyes =agedata['design_comp'] == 'Yes'
agedata.loc[condyes,'Yes'] = 1

condno =agedata['design_comp'] == 'No'
agedata.loc[condno,'No'] = 1

#append personalised_price
agedata["under 500"] = 0
agedata["0"] = 0
agedata["500-1000"] = 0
agedata["over 1000"] = 0

condu500 =agedata['personalised_price'] == 'under 500'
agedata.loc[condu500,'under 500'] = 1

cond0 =agedata['personalised_price'] == '0'
agedata.loc[cond0,'0'] = 1

cond5001000 =agedata['personalised_price'] == '500-1000'
agedata.loc[cond5001000,'500-1000'] = 1

condo1000 =agedata['personalised_price'] == 'over 1000'
agedata.loc[condo1000,'over 1000'] = 1

#create dataframes for the other demographics
genderdata = pd.DataFrame(agedata[['gender', 'only pro', 'Yes', 'No', 'under 500', '0', '500-1000', 'over 1000']])
owncardata = pd.DataFrame(agedata[['own_car', 'only pro', 'Yes', 'No', 'under 500', '0', '500-1000', 'over 1000']])
rsdata = pd.DataFrame(agedata[['relationship', 'only pro', 'Yes', 'No', 'under 500', '0', '500-1000', 'over 1000']])

agedata = agedata.drop(['age', 'gender', 'own_car', 'relationship','design_comp', 'personalised_price', '3D_Design'], axis=1)
agedata
```

```python
frequent_itemsets = apriori(agedata, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules[(rules['confidence'] >= 0.5)]
```

```python
#repeat for gender
#remove anomaly of prefer not to say
genderdata = genderdata.drop(genderdata[genderdata['gender']=="Prefer not to say"].index)

#append gender categories
genderdata["Male"] = 0
genderdata["Female"] = 0


condmale =genderdata['gender'] == 'Male'
genderdata.loc[condmale,'Male'] = 1

condfemale =genderdata['gender'] == 'Female'
genderdata.loc[condfemale,'Female'] = 1
genderdata = genderdata.drop(['gender'], axis=1)
genderdata
```

```python
frequent_itemsets_gender = apriori(genderdata, min_support=0.07, use_colnames=True)
rules_gender = association_rules(frequent_itemsets_gender, metric="lift", min_threshold=1)
rules_gender[(rules_gender['confidence'] >= 0.5)]
```

```python
#append own_car categories
owncardata["No Car"] = 0
owncardata["1 Car"] = 0
owncardata["More than 1 Car"] = 0


condnocar =owncardata['own_car'] == 'Do not own a car, but planning to purchase in future'
owncardata.loc[condnocar,'No Car'] = 1

cond1car =owncardata['own_car'] == 'Own a car'
owncardata.loc[cond1car,'1 Car'] = 1

condmorecar =owncardata['own_car'] == 'Own more than one car'
owncardata.loc[condmorecar,'More than 1 Car'] = 1
owncardata = owncardata.drop(['own_car'], axis=1)
owncardata
```

```python
frequent_itemsets_owncar = apriori(owncardata, min_support=0.1, use_colnames=True)
rules_owncar = association_rules(frequent_itemsets_owncar, metric="lift", min_threshold=1)
rules_owncar[(rules_owncar['confidence'] >= 0.5)]
#(rules_owncar['lift'] >= 1) &
```

```python
#For relationship
#append own_car categories
rsdata["No Children"] = 0
rsdata["Children"] = 0
rsdata["Single"] = 0


condsingle =rsdata['relationship'] == 'Single'
rsdata.loc[condsingle,'Single'] = 1

condnochild =rsdata['relationship'] == 'Married with no children'
rsdata.loc[condnochild,'No Children'] = 1

condchild =rsdata['relationship'] == 'Married with children'
rsdata.loc[condchild,'Children'] = 1
rsdata = rsdata.drop(['relationship'], axis=1)
rsdata
```

```python
frequent_itemsets_rs = apriori(rsdata, min_support=0.1, use_colnames=True)
rules_rs = association_rules(frequent_itemsets_rs, metric="lift", min_threshold=1)
rules_rs[(rules_rs['confidence'] >= 0.5)]
```

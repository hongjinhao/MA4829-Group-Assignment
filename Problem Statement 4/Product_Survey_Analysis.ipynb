## Clean up the buyer_factor, int_components and ext_components columns to have a unqiue column for each option

```python
# Step 1: Identify all unique factors
unique_factors = set()
for factors in car_data['buyer_factor'].dropna():
    unique_factors.update(factors.split(';'))
unique_factors
```

```python
# Step 2: Create new columns for each factor and initialize to 0
for factor in unique_factors:
    car_data[factor] = 0
```

```python
# Step 3: Set the value to 1 if the factor is present for that row
for index, row in car_data.iterrows():
    if pd.notna(row['buyer_factor']):
        factors = row['buyer_factor'].split(';')
        for factor in factors:
            if factor in unique_factors:
                car_data.at[index, factor] = 1
```

```python
# Step 1: Identify all unique factors
unique_factors = set()
for factors in car_data['ext_components'].dropna():
    unique_factors.update(factors.split(';'))
unique_factors
```

```python
# Step 2: Create new columns for each factor and initialize to 0
for factor in unique_factors:
    car_data[factor] = 0
# Step 3: Set the value to 1 if the factor is present for that row
for index, row in car_data.iterrows():
    if pd.notna(row['ext_components']):
        factors = row['ext_components'].split(';')
        for factor in factors:
            if factor in unique_factors:
                car_data.at[index, factor] = 1
```

```python
# Step 1: Identify all unique factors
unique_factors = set()
for factors in car_data['int_components'].dropna():
    unique_factors.update(factors.split(';'))
unique_factors
```

```python
# Step 2: Create new columns for each factor and initialize to 0
for factor in unique_factors:
    car_data[factor] = 0
# Step 3: Set the value to 1 if the factor is present for that row
for index, row in car_data.iterrows():
    if pd.notna(row['int_components']):
        factors = row['int_components'].split(';')
        for factor in factors:
            if factor in unique_factors:
                car_data.at[index, factor] = 1
```

```python
for column in car_data.columns:
    sb.catplot(y=column, data=car_data, kind="count")
    plt.title(f'Count Plot for {column}')
    plt.show()
```
